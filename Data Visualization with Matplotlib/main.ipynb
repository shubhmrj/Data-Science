```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
data = pd.read_csv("google_cleaned.csv",index_col=0,header=0)
# so i give index_col = 0 because here is some abguity in given data , like indexing show in data is two first by default of pandas and second of given by dataset
```

### Visualisation

LIne PLOT

```python
plt.plot(data['Apps'],data['Rating'])
plt.xlabel('APPS')
plt.ylabel('Ratings')
plt.title("Apps And its Rating")
```

```python
plt.scatter(data['Apps'],data['Rating'])
plt.xlabel('APPS')
plt.ylabel('Ratings')
plt.title("Apps And its Rating")
```

```python
# so the due large data set , there may be dificullty in visualize the data

plt.scatter(
    data['Apps'],
    data['Rating'],
    alpha=0.05
)

# still slow in millions
```

```python
sampled = data.sample(n=5000, random_state=42)
plt.scatter(sampled['Apps'], sampled['Rating'])
```

```python
plt.scatter(
    data['Apps'],
    data['Rating'],
    alpha=0.05
)
plt.xscale('log') 
```

```python
plt.hexbin(
    data['Apps'],
    data['Rating'],
    gridsize=50
)
plt.colorbar(label='count')
```

```python
plt.hist2d(
    data['Apps'],
    data['Rating'],
    bins=50
)
plt.colorbar()
```

```python
plt.figure(figsize=(10,15))
plt.hexbin(
    data['Price'],
    data['Rating'],
    gridsize=60
)
plt.xscale('log')
plt.colorbar()
plt.show()
```
