```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,1000)
```

```python
x
```

```python
y = np.sqrt(10**2 - x**2)
```

```python
y
```

```python
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

# Inner Cycle

```python
x1 = np.linspace(-5.0, 5.0, 100)
y1 = np.sqrt(5**2 - x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

# Visualize

```python
plt.scatter(x,y)
plt.scatter(x1,y1)
```
