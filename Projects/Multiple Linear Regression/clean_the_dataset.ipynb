```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
# conversion as demanded by stakeholders
df['total_rom'] = df["total_rooms"]/df['households']
```

```python
df.head()
```

```python
df.rename(columns={'total_rom':'average_rooms'},inplace=True)
```

```python
df.head()
```

```python
# conversion as demanded by stakeholders

df['Average_Bedrooms'] = df["total_bedrooms"] / df['households']
df['Average_Population_Home'] = df["population"] / df['households']
df["Price"]     = df["median_house_value"] / 100000
```

```python
df.head(10)
```

```python
df.drop(columns=['total_rooms','total_bedrooms','median_house_value'],inplace=True)
```

```python
df.head(10)
```

```python
df = df[
    [
        "median_income", "housing_median_age", "average_rooms", "Average_Bedrooms",
        "population", "Average_Population_Home", "latitude", "longitude", "ocean_proximity", "Price"
    ]
]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```
