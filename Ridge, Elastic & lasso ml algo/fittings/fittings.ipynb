```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
```

```python
# create datasets
x= np.linspace(-3,3,20).reshape(-1,1)
y=0.5*x**2+1.5*x+2+np.random.randn(20,1)
```

```python
plt.plot(x,y)
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
plt.figure(figsize=(10,2))
```

```python
plt.scatter(x,y,colorizer='g',marker='o',s=20,linewidth=2,alpha=0.5,zorder=2,cmap='viridis')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
plt.figure(figsize=(10,2))
```

```python
degrees=[1,2,15]

x_plot=np.linspace(-3,3,200).reshape(-1,1)
```

```python
x_plot
```

```python
for degree in degrees:
    model=Pipeline([
        ("poly_features",PolynomialFeatures(degree=degree,include_bias=False)),
        ("lin_reg",LinearRegression())
    ])
    model.fit(x,y)
    y_pred = model.predict(x_plot)
    plt.plot(x_plot, y_pred, label=f"Degree {degree}", linewidth=2)
    plt.legend()
```

```python
plt.scatter(x, y, color="black", label="Training Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Underfitting vs Good Fit vs Overfitting")
plt.show()
```

```python
degrees = [1, 2, 15]
plt.figure(figsize=(10, 6))

x_plot = np.linspace(-3, 3, 200).reshape(-1, 1)

for degree in degrees:
    model = Pipeline([
        ("poly_features", PolynomialFeatures(degree=degree, include_bias=False)),
        ("lin_reg", LinearRegression())
    ])
    model.fit(x, y)
    y_pred = model.predict(x_plot)
    plt.plot(x_plot, y_pred,label=f"Degree {degree}",  linewidth=2)

plt.scatter(x, y, color="black", label="Training Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Underfitting vs Good Fit vs Overfitting")
plt.show()
```
