```python
# train test split

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
def coorelation(dataset, threshold):
    col_corr = set()
    col_matrix= dataset.corr()
    
    for i in range(len(col_matrix.columns)):
        for j in range(i):
            if abs(col_matrix.iloc[i,j] > threshold):
                colname= col_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=coorelation(X_train,0.85)
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train_scaled= scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
# box plot
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_tain before Scaling')

plt.subplot(1,2,2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')

```

## linear regresion


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lin_reg= LinearRegression()

lin_reg.fit(X_train_scaled,y_train)

y_predication = lin_reg.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_predication)

score = r2_score(y_test,y_predication)

print('Mean Absolute error',mae)
print("R2_Score", score)

```

```python
plt.scatter(y_test,y_predication)
```

## lasso



```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
lasso= Lasso()

lasso.fit(X_train_scaled,y_train)

y_predication = lasso.predict(X_test_scaled)

mae = mean_absolute_error(y_test,y_predication)

score = r2_score(y_test,y_predication)

print('Mean Absolute error',mae)
print("R2_Score", score)

```

```python
plt.scatter(y_test,y_predication)
```

## Ridge
