```python
import sqlite3 as db
```

```python
# connect to database

connection = db.connect('sql.db')
```

```python
# create a cursor to iterate the table

cursor = connection.cursor()
```

```python
# create a table

cursor.execute('''
              CREATE TABLE IF NOT EXISTS users(
                  id INTEGER PRIMARY KEY AUTOINCREMENT,
                  name TEXT not null,
                  age INTEGER,
                  email UNIQUE
              ) 
''')

connection.commit()
```

```python
# interstion

cursor.execute(
    "INSERT INTO users (name,age,email) VALUES(?, ?, ?)",('SHUBHAM',23,'SRNWDAWA')
)
connection.commit()
```

```python
users= [
    ('raj',23,'prnwda'),
    ('rani',24,'pwda')
]

cursor.executemany(
    "INSERT INTO users (name,age,email) VALUES(?, ?, ?)",users
)
connection.commit()
```

```python
cursor.execute("SELECT * FROM users")
rows=cursor.fetchall()

for row in rows:
    print(row)
```

```python
# Update
cursor.execute(
    "UPDATE users set age=? WHERE id=?",
    (28,2)
)
connection.commit()
```
