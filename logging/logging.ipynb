```python
import logging
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%y-%m-%d %H:%M:%S'
    )

logging.debug("This is the debug message")
logging.critical("this is the critical message")

```

```python
import logging

logger = logging.getLogger()
logger.setLevel(logging.DEBUG)

file_handler = logging.FileHandler("app.log", mode="w")
formatter = logging.Formatter(
    "%(asctime)s-%(name)s-%(levelname)s-%(message)s",
    "%y-%m-%d %H:%M:%S"
)

file_handler.setFormatter(formatter)
logger.addHandler(file_handler)

logger.debug("This is the debug message")
logger.critical("this is the critical message")
```

```python
# multiple logger
import logging 

logger_1=logging.getLogger("module_1")
logger_1.setLevel(logging.WARNING)

logger_2=logging.getLogger("module_2")
logger_2.setLevel(logging.DEBUG)

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%y-%m-%d %H:%M:%S'
    )
```
