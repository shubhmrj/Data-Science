```python
corpus =  """Hey, my name is shubham raj.
I am currently btech in computer science from malwa institute of technology!
currenlt i am work on my project.
        
"""
```

```python
corpus
```

```python
print(corpus)
```

```python
##  Tokenization
## Sentence-->paragraphs

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
documents
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization 
## Paragraph-->words
## sentence--->words

from nltk.tokenize import word_tokenize
```

```python
word_document = word_tokenize(corpus)
```

```python
word_document
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
for sentence in word_document:
    print(sent_tokenize(sentence))
```

```python
for sentence in word_document:
    print(sentence)
```
