    Named Entity Recognition (NER) in NLP identifies and categorizes key information in text into predefined classes, transforming unstructured data into structured insights. Standard categories include Persons, Organizations, Locations (GPE), Dates, Times, Percentages, and Monetary values. Common tools include SpaCy and NLTK. 


## Common NER Entity Types

### NER systems commonly recognize the following entities: 

    PERSON (PER): Names of people, including fictional characters.
    ORGANIZATION (ORG): Companies, agencies, institutions (e.g., Google, United Nations).
    LOCATION (LOC/GPE): Geographical locations, countries, cities, states.
    DATE/TIME: Specific calendar dates or durations.
    MONEY: Monetary values, including unit.
    PERCENT (PERCENTAGE): Percentage values.
    FACILITY (FAC): Buildings, airports, highways.
    PRODUCT: Tangible objects, vehicles, foods, software.
    EVENT: Named hurricanes, battles, wars, sports events.
    WORK_OF_ART: Titles of books, songs, movies.
    LANGUAGE: Named languages.
    NORP: Nationalities, religious groups, and political groups. 
    Common NER Models and Approaches
    Rule-Based: Uses dictionaries and regular expressions to identify entities.
    Machine Learning (Traditional): Uses algorithms like Conditional Random Fields (CRF) or Support Vector Machines (SVM).
    Deep Learning: Utilizes models such as BiLSTM-CRF for sequence labeling.
    Transformers: BERT-based models provide state-of-the-art performance by understanding context. 

### Key Libraries and Tools
- spaCy: Known for high performance, accuracy, and ease of use in production.
- NLTK: Commonly used for education and research.
- Stanford CoreNLP: Offers robust pre-trained models for various entity types.
- Hugging Face Transformers: Popular for utilizing state-of-the-art BERT, RoBERTa, and other transformer models

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
from nltk import word_tokenize
```

```python
words = word_tokenize(sentence)
```

```python
import nltk
tag_elements=nltk.pos_tag(words)
```
